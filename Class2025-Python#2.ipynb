```python
# Example : ABC Company has reported monthly revenue and cost as below:
revenue = [123, 5, 33, 10, 12, 20, 40, 1.2, 9.9, 0.5, 11.11, 999]
cost = [10, 15, 100, 10, 5, 2, 13.1, 10, 99, 0, 3.2, 555]
```

```python
# Calculate monthly profit of ABC Company

revenue = np.array(revenue)
cost = np.array(cost)
profit = revenue - cost
```

```python
print(profit)
```

```python
# Calculate Total revenue, Total cost and Total profit of the company
# Hint : Using np.sum()
Total_revenue = np.sum(revenue)
Total_cost = np.sum(cost)
Total_profit = np.sum(profit)
```

```python
print(Total_revenue)
print(Total_cost)
print(Total_profit)
```

```python
# Calculating Average, Median and Standart Deviation of the revenue
print(np.mean(revenue))
print(np.median(revenue))
print(np.std(revenue))
```

```python
# Calculating Correlation of reveue and cost
np.corrcoef(revenue,cost)
```

## **Slicing**

- Basic : dataset[start : until + 1]
- Advannce : dataset[start : until +1 : every n th]

```python
# Revenue
revenue
```

```python
# Calculate Quarterly revenue as revenue_Q1, ... , revenue_Q4
revenue_Q1 = revenue[:3]
revenue_Q2 = revenue[3:6]
revenue_Q3 = revenue[4:9]
revenue_Q4 = revenue[9:]
print(revenue_Q1)
print(revenue_Q2)
print(revenue_Q3)
print(revenue_Q4)
```

```python
# Calculate Quarterly cost as cost_Q1, ... , coat_Q4
cost_Q1 = cost[:3]
cost_Q2 = cost[3:6]
cost_Q3 = cost[4:9]
cost_Q4 = cost[9:]
print(cost_Q1)
print(cost_Q2)
print(cost_Q3)
print(cost_Q4)
```

```python
# Calculate Quarterly profit as profit_Q1, ... profit_Q4
profit_Q1 = profit[:3]
profit_Q2 = profit[3:6]
profit_Q3 = profit[4:9]
profit_Q4 = profit[9:]
print(profit_Q1)
print(profit_Q2)
print(profit_Q3)
print(profit_Q4)
```

```python
# Show revenue from month 1,3,5,...,11 (index = 0,2,4,...,10)
revenue[0:11:2]
```

### **TVOM**
Time value of Money

```python
# Assigning variables

pv = 100
r = 0.05
n = 10

fv = pv * (1 + r)**n
print(round(fv, 2))
```

```python
# Continuos Compounting

fv_c = pv * np.exp(r * n)
print(round(fv_c, 2))
```
